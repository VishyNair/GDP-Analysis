Cleaning the datasets by removing unwanted columns which are not concerning our current analysis and to impute values if needed.

````python
#To remove the 'GSDP CURRENT PRICES' AND '% GROWTH OVER PREVIOUS YEAR' ROWS FOR 2016-17
df = df.loc[df['Duration'] != '2016-17',:]
#TO remove all rows with 'GSDP CURRENT PRICES' for other years and only to retain '% GROWTH OVER PREVIOUS YEAR' rows.
new_df = df.loc[((df['Items  Description'] == '(% Growth over previous year)')) & ((df['Duration'] =='2013-14')|(df['Duration'] =='2014-15')|(df['Duration'] =='2015-16'))]
````

````python
#Dropping the Items description column which is not needed here.
new_df = new_df.drop('Items  Description',axis=1)
#Dropping the West Bengal1 column, since the data is unavailable for all the year periods.
new_df = new_df.drop('West Bengal1',axis=1)
````

````python
#Viewing the current dataframe
new_df
````

````python
#Transposing the rows and columns to get it ready for visualisation
GDP_df = new_df.T
GDP_df
````

````python
#new_df_transposed['States']=new_df_transposed.index
````

````python
#Renaming the column names to get the data correct for each year
GDP_df.rename(columns={7:'2013-14'},inplace=True)
GDP_df.rename(columns={8:'2014-15'},inplace=True)
GDP_df.rename(columns={9:'2015-16'},inplace=True)
````

````python
#To remove the duration row which is unwanted now.
GDP_df = GDP_df.iloc[1:, ]
GDP_df
````

````python
#To introduce a new column 'states' from the index.
GDP_df['States']=GDP_df.index
````

````python
#To set the new column 'States' as index so as to get the column index correct.
GDP_df.set_index('States',inplace=True)
GDP_df
````

````python
#Imputing the missing values for the missing year for different states.
GDP_df.loc['Himachal Pradesh','2015-16'] = GDP_df.loc['Himachal Pradesh'].mean()
GDP_df.loc['Maharashtra','2015-16'] = GDP_df.loc['Maharashtra'].mean()
GDP_df.loc['Manipur','2015-16'] = GDP_df.loc['Manipur'].mean()
GDP_df.loc['Mizoram','2015-16'] = GDP_df.loc['Mizoram'].mean()
GDP_df.loc['Nagaland','2015-16'] = GDP_df.loc['Nagaland'].mean()
GDP_df.loc['Punjab','2015-16'] = GDP_df.loc['Punjab'].mean()
GDP_df.loc['Rajasthan','2015-16'] = GDP_df.loc['Rajasthan'].mean()
GDP_df.loc['Tripura','2015-16'] = GDP_df.loc['Tripura'].mean()
GDP_df.loc['Andaman & Nicobar Islands','2015-16'] = GDP_df.loc['Andaman & Nicobar Islands'].mean()
GDP_df
````

````python
#To introduce a new column 'Mean' for the mean of the dealing years
GDP_df['Mean'] = (GDP_df['2013-14'] + GDP_df['2014-15'] + GDP_df['2015-16'])/3
````

````python
#To get the state names as new column from the index which will help while plotting.
GDP_df['States'] = GDP_df.index
GDP_df
````

````python
#Dropping the current index and resetting
GDP_df = GDP_df.reset_index(drop=True)
GDP_df
````

````python
#Sorting the GDP dataframe in descending order of the Mean value so that it will help in visualisation.
GDP_df = GDP_df.sort_values(by='Mean',ascending = False)
GDP_df
````

````python
# To plot the mean of the Percentage growth of the previous years against the states.
plt.figure(figsize=(12,8))
sns.barplot(y = 'States',x = 'Mean',data=GDP_df,color=(0.2, 0.4, 0.6, 0.6))
plt.xlabel('Average of percentage growth from 2013-14 to 2015-16')
plt.show()
````

From the above plot we could easily find out ```Mizoram``` is the leading state which has been growing consistently faster compared to other states followed by ```Tripura``` and ```Meghalaya```.
At the bottom comes```Odisha``` , ```Meghalaya``` and ```Goa``` which are the states struggling.

## Part-II: GDP and Education Drop-out Rates

````python
#Reading the 'State-wise Average Annual Drop-out rates from 2012-13 to 2014-15' csv file to the dataframe df.
drop_out_df = pd.read_csv('./Data-II/rs_session243_au570_1.1.csv')
````

````python
#To view the drop out dataframe.
drop_out_df
````

````python
#Sorting the 'drop_out_df' dataframe according to state.
drop_out_df.sort_values(by='Level of Education - State',ascending= True)
````

````python
#Sorting the all_states dataframe by 'States'.
all_states.sort_values(by='States',ascending= True)
````

````python
#Merging the 'drop_out_df' and the 'all_states' dataframe. Here the dataframes are joined based on states 
#Inner join is used here to get the details where state names are present in both the dataframes.
#All the UT's as well as details of Jammu & Kashmir and West Bengal is dropped since it is not present in all_states.
drop_out = pd.merge(drop_out_df,all_states, how="inner", left_on="Level of Education - State", right_on="States")
````

````python
#To view the head of the new dataframe 'drop_out'
drop_out.head()
````

````python
#Dropping the unnecessary columns in the mergd dataframe
drop_out = drop_out.drop(['Primary - 2012-2013','Primary - 2014-2015.1','Upper Primary - 2012-2013','Upper Primary - 2013-2014','Secondary - 2012-2013','Secondary - 2013-2014','Senior Secondary - 2012-2013','Senior Secondary - 2013-2014','States','Sl. No.','Category'],axis=1)
````

````python
#Imputing all the missing values in the dataframe with 0
drop_out.fillna(0,inplace=True)
````

````python
#Subsetting the Percapita GDSP and the primary,upper primary,Secondary,Senior Secondary drop out percentages to 'drop_out_subset' dataframe
drop_out_subset = drop_out[["Per Capita GSDP (Rs.)","Primary - 2014-2015","Upper Primary - 2014-2015","Secondary - 2014-2015","Senior Secondary - 2014-2015"]]
````

````python
#Viewing the new dataframe
drop_out_subset
````

````python
#Plotting the pairplot as well for the 'drop_out_subset'
sns.pairplot(drop_out_subset)
plt.show()
````

````python
#Finding the correlation matrix for 'drop_out_subset'
drop_out_cor = drop_out_subset.corr()
round(drop_out_cor, 3)
````

````python
#Plotting the drop_out_cor correlation matrix as heatmap
# figure size
plt.figure(figsize=(10,8))
# heatmap
sns.heatmap(drop_out_cor, cmap="Blues", annot=True)
plt.show()
````

So from the above heatmap for the correlation matrix between the per capita GDSP for the states and the different sections of education, it is clear that there is comparative high negative correlation between the per capita GDSP and the upper primary dropout percentages. That means the per capita GDP is getting low whenever the drop out percentages is more in the upper primary section.

So the above inference will help directly for all states to give more importance to the upper primary education for all the children in that age group.

Followed by the upper primary is the primary and secondary dropouts which have negative correlation with the per capita GDSP. So more the primary and secondary dropouts lesser the per capita GDSP it is.
